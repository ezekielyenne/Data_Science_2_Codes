```python
# Defining the field in the dataframe
d={'Name':['John',' Ruth', 'Lydia',' Joseph', 'Sam', 'James', 'Easy', 'Tall', 'Mamis', 'Ado'],
   'Subjects':['Chemistry', 'Maths', 'English', 'Physics', 'Biology', 'English', 'Physics', 'Biology', 'Chemistry', 'Maths'],
   'Score_1':[7,6,6,6,7,8,9,9,8,6],
  'Score_2':[6,5,5,7,8,3,9,8,9,5]}
# Creating the data frame using Pandas. The data frame is called df
df=pd.DataFrame(d)

print(df)
```

```python
# Addition of percentage field to the frame using total scores as Percentage_score = scores / total scores * 100

df['percentage_%']=((df['Score_1']+ df['Score_2'])/20)*100

# Print the data frame with the added column
print(df)
```

```python
# Select only the percentage column and rename usign df1
df1=df['percentage_%']

# shcek the new column df1
print(df1)
```

```python
# Defining a function that will add grade to the student scores

for i in range(len(df1)):          # Using i to count for the number of characters in range(len(df))

 # check if the score is greater than 80 and print grade A
 if(df1.iloc[i]>80):
  print('A')
 # Check if then score is greater than or equal to 60 and print B
 elif (df1.iloc[i]>=60):
  print('B')
# Otherwise print fail as the grade
 else:
  print('FAIL')
```
